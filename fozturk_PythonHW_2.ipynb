The properties of materials can be changed by adding new atoms known as doping. Figure 2 shows the variation of the electric coefficient with the frequency of the applied electric field as a result of doping the Zinc oxide (ZnO) substrate with different percentages of Boron (B) or nitrogen (N) nanoparticles.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel (r'ZNO_data_fig2.xlsx', names=['Freq','ZNO_Eps','ZNO_N4_Eps','ZNO_B4_Eps','ZNO_N2_Eps','ZNO_B2_Eps'])
```

```python
df.head()
```

```python
plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"
plt.rcParams['axes.linewidth'] = 2 
plt.figure(figsize=(12, 10))
plt.plot(df.Freq, df.ZNO_Eps, "-k", marker="s",label="ZNO")
plt.plot(df.Freq, df.ZNO_B2_Eps, "-r", marker="o", label="ZNO_B2")
plt.plot(df.Freq, df.ZNO_B4_Eps, "-b", marker=">", label="ZNO_B4")
plt.plot(df.Freq, df.ZNO_N2_Eps, "-g", marker="^", label="ZNO_N2")
plt.plot(df.Freq, df.ZNO_N4_Eps, "Magenta", marker="D", label="ZNO_N4")
plt.yscale("log")
plt.xscale("log")
plt.xlabel("f (Hz)", fontsize=22)
plt.ylabel('$\\epsilon^\prime$', fontsize=22)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.axis([10**-2, 10**7, 10**0, 3*10**3])
plt.legend(loc="center right",fontsize=20)
plt.savefig('ZNO_freq_vs_eps.pdf')
```
